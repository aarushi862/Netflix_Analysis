```python
df.drop(columns=["director"],inplace=True)
```

```python
df['cast'].fillna(df['cast'].mode()[0],inplace=True)
df['country'].fillna(df['country'].mode()[0],inplace=True)
```

```python
df['date_added'].fillna(df['date_added'].mode()[0],inplace=True)
```

```python
df['rating'].fillna(df['rating'].mode()[0],inplace=True)
```

```python
df['duration'].fillna(df['duration'].mode()[0],inplace=True)
```

```python
df.isnull().sum()
```

```python
df.columns=df.columns.str.lower()
df.columns=df.columns.str.replace(' ','_')
```

```python
df['type'].value_counts()
```

```python
plt.figure(figsize=(5,4))
sns.set_style("whitegrid")
sns.countplot(x='type',data=df,palette='pastel')
plt.xlabel('Type')
plt.ylabel('Number of content')
plt.title('Movies vs Tv Shows')
plt.show()
```

```python
df['listed_in'].value_counts().head(10)
```

```python
data=df['release_year'].value_counts().sort_index()
```

```python
plt.figure(figsize=(5,5))
sns.lineplot(x=data.index , y=data.values)
plt.xlabel('Release year')
plt.ylabel('Number of content')
plt.title('Content release over years')
plt.show()
```

```python
data1=df['country'].value_counts().head(10)
```

```python
plt.figure(figsize=(11,5))
sns.set_style("whitegrid")
sns.barplot(x=data1.index , y=data1.values,palette="viridis")
plt.xlabel('Country')
plt.ylabel('Number of content')
plt.title('Top 10 country')
```

```python
data2=df['rating'].value_counts()
```

```python
plt.figure(figsize=(5,5))
plt.xticks(rotation=90)
sns.set_style("whitegrid")
sns.barplot(x=data2.index , y=data2.values,palette='Set2')
plt.yticks(np.arange(0, max(data2.values)+300, 300))
plt.xlabel('Rating')
plt.ylabel('Number of content')
plt.title('Rating')
plt.show()
```

```python
data3=df['duration'].value_counts().head(10)
```

```python
plt.figure(figsize=(9,5))
sns.set_style("whitegrid")
sns.barplot(x=data3.index , y=data3.values,palette='coolwarm')
plt.yticks(np.arange(0, max(data3.values)+200, 200))
for i, v in enumerate(data3.values):
    plt.text(i, v + 10, str(v), ha='center')
plt.xlabel('Duration')
plt.ylabel('Number of content')
plt.title('Top 10 Longest Duration Content')
plt.show()
```

```python
df['type'].value_counts()
```

```python
df_movie=df[df['type']=='Movie']
df_tv=df[df['type']=='TV Show']
```

```python
movie_rating=df_movie['rating'].value_counts()
tv_rating=df_tv['rating'].value_counts()
```

```python
movie_top=movie_rating.head(10)
tv_top=tv_rating.head(10)
```

```python
plt.figure(figsize=(9,5))
# movies
plt.subplot(1,2,1)
movie_top.plot(kind='bar',color='orange')
plt.title('Top ratings Movies')
plt.xticks(rotation=90)

for i,v in enumerate(movie_top.values):
    plt.text(i,v+10,str(v),ha='center')
#tv shows
plt.subplot(1,2,2)
tv_top.plot(kind='bar',color='skyblue')
plt.title('Top ratings Tv Shows')
plt.xticks(rotation=90)

for i, v in enumerate(tv_top.values):
    plt.text(i,v+10, str(v), ha='center')

plt.tight_layout()
plt.show()

```
